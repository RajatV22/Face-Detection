```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
plt.imshow(img[...,::-1]) #this conversion can be performed inline!
plt.show()
```

```python
img_g = cv2.GaussianBlur(img,(15,15),7) #use a 15x15 Gaussial kernel with standard deviation 7
plt.imshow(img_g[...,::-1])
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#print (img_gray.shape) #now the image has just one channel
plt.imshow(img_gray[...,:],cmap='gray') #cmap='gray' is required to show the image correctly
```

```python
img_canny = cv2.Canny(img,180,180)
plt.imshow(img_canny[...,::-1],cmap='gray')
```

```python
import cv2
cap = cv2.VideoCapture(0) #open camera with device "0"

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read() # return value "ret", and set to True if there are no errors
    # Our operations on the frame come here
    #gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    cv2.imshow('frame',frame)
    if cv2.waitKey(40) & 0xFF == ord('q'): #wait for 40 ms (i.e., 25 fps or key q
        break
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
